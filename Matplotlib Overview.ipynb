```python
%matplotlib inline
import numpy as np
import pandas as pd
```

<h4>Read the csv file of your choice</h4>

```python
patients = pd.read_csv("MOCK_PATIENT_DATA.csv")
```

<h4>Plot counts of a specified column using Pandas</h4>

```python
fig = patients.residence_state.value_counts()[:5].plot(kind='barh')
```

<h4>Bar plot of median values</h4>

```python
diagnosis_info = patients.groupby('diagnosis_category').agg(np.median)
del diagnosis_info['id']
#['diagnosis_age'].agg(np.median)
diagnosis_info.index = ['Infections','Parasitic','Neoplasms','Blood Disease',
                    'Endocrine', 'Mental','Nervous System','Eye','Ear','Circulatory',
                 'Respiratory','Digestive','Skin','Musculoskeletal','Genitourinary',
                   'Pregnancy','Perinatal','Congenital','Not_Classified', 'Injury',
                    'External Causes','Health Services', 'Deleted', 'Expanded']
diagnosis_info.plot(kind = 'bar', title='Median')
#patients.groupby('diagnosis_category')['diagnosis_age'].agg(np.median).plot(kind='bar')
```
